```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_df=df['quality']
x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.2)
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.3, random_state=17)
print('Lengths:', len(x_train), len(x_valid), len(x_test))
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor, Ridge, Lasso, RidgeCV, LassoCV
```

```python
def check_regression(r, x_v, y_v):
    y_p = r.predict(x_v)
    return mean_squared_error(y_v, y_p), mean_absolute_error(y_v, y_p), r2_score(y_v, y_p)
```

```python
print('Check the Linear regression model at the default parameters')
l_reg = LinearRegression()
l_reg = l_reg.fit(x_train, y_train)
mse, mae, r2 = check_regression(l_reg, x_valid, y_valid)
print(f'Linear regression model validation metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
```

```python
mse, mae, r2 = check_regression(l_reg, x_test, y_test)
print(f'Linear regression model test metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
```

```python
print('Check the SGD regression model at the default parameters')
s_reg = SGDRegressor(max_iter=1000, tol=1e-3)
s_reg = s_reg.fit(x_train, y_train)
mse, mae, r2 = check_regression(s_reg, x_valid, y_valid)
print(f'SGD regression model validation metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
```

```python
mse, mae, r2 = check_regression(s_reg, x_test, y_test)
print(f'SGD regression model test metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
```

```python
print('Check the Ridge regression model at the default parameters')
r_reg = Ridge()
r_reg = r_reg.fit(x_train, y_train)
mse, mae, r2 = check_regression(r_reg, x_valid, y_valid)
print(f'Ridge regression model validation metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
```

```python
mse, mae, r2 = check_regression(r_reg, x_test, y_test)
print(f'Ridge regression model test metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
```

```python
print('Get the best regularization parameter Alpha (an easy way)')
rcv_reg = RidgeCV(alphas=np.logspace(-6, 6, 13))
rcv_reg = rcv_reg.fit(x_train, y_train)
mse, mae, r2 = check_regression(rcv_reg, x_valid, y_valid)
print(f'Ridge regression model test metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
print('RidgeCV alpha', rcv_reg.alpha_)
```

```python
print('Try to apply the Alpha')
r_reg = Ridge(alpha=0.1)
r_reg = r_reg.fit(x_train, y_train)
mse, mae, r2 = check_regression(r_reg, x_valid, y_valid)
print(f'Ridge regression model validation metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
```

```python
mse, mae, r2 = check_regression(r_reg, x_test, y_test)
print(f'Ridge regression model test metrics: MSE:{mse} MAE: {mae} R^2 score {r2}')
```

```python
print('The result: the best model from above mentioned ones by MSE score: Ridge(alpha=0.1)')
```
