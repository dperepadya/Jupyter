```python
print('Remove the Target column')
df = df.drop(['kredit'], axis=1)
```

```python
df.head(5)
```

```python
print("Dataset features normalization")
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_cols = df.select_dtypes(include='number').columns
norm_df = df
for c in num_cols:
    norm_df.loc[:, c] = scaler.fit_transform(norm_df[[c]])
```

```python
print('Correlation matrix (numeric features and metrics)')
corr_matrix = norm_df.corr()
sns.heatmap(corr_matrix);
```

```python
print('PairPlots')
%config InlineBackend.figure_format = 'png'
sns.pairplot(norm_df);
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
import matplotlib.pyplot as plt
plt.scatter(norm_df.iloc[:, 0], norm_df.iloc[:, 1])
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
kmeans = KMeans(n_clusters=8, random_state=17, n_init="auto")
model = kmeans.fit(norm_df)
predicted = kmeans.fit_predict(norm_df)
n_clusters_ = len(set(predicted)) - (1 if -1 in predicted else 0)
n_noise_ = list(predicted).count(-1)
print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
silhouette = silhouette_score(norm_df, predicted)
davies_bouldin = davies_bouldin_score(norm_df, predicted)
calinski_harabasz = calinski_harabasz_score(norm_df, predicted)
print("Silhouette score: ", silhouette)
print("Davies-Bouldin score: ", davies_bouldin)
print("Calinski-Harabasz score function: ", calinski_harabasz)
```

```python
plt.scatter(norm_df.iloc[:, 0], norm_df.iloc[:, 1], c=predicted)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', s=200)
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1, min_samples=8)
model = dbscan.fit(norm_df)
labels = model.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
silhouette = silhouette_score(norm_df, predicted)
davies_bouldin = davies_bouldin_score(norm_df, predicted)
calinski_harabasz = calinski_harabasz_score(norm_df, predicted)
print("Silhouette score: ", silhouette)
print("Davies-Bouldin score: ", davies_bouldin)
print("Calinski-Harabasz score function: ", calinski_harabasz)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.7)
pca.fit(norm_df)
reduced_df = pca.transform(norm_df)
```

```python
%matplotlib notebook
#import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (6,4)

fig, ax = plt.subplots()
xi = np.arange(0, 8, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 8, step=1))
plt.ylabel('Cumulative variance (%)')
plt.title('Information Loss Diagramm')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=12)

ax.grid(axis='x')
plt.show()
```

```python
kmeans = KMeans(n_clusters=8, random_state=17, n_init="auto")
predicted = kmeans.fit_predict(reduced_df)
silhouette = silhouette_score(reduced_df, predicted)
davies_bouldin = davies_bouldin_score(reduced_df, predicted)
calinski_harabasz = calinski_harabasz_score(reduced_df, predicted)
print("Silhouette score: ", silhouette)
print("Davies-Bouldin score: ", davies_bouldin)
print("Calinski-Harabasz score function: ", calinski_harabasz)
```

```python
fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = (6,4)
plt.scatter(norm_df.iloc[:, 0], norm_df.iloc[:, 1], c=predicted)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', s=200)
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
